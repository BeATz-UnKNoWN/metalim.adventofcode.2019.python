```python
import intcode
from itertools import count

with open('inputs/19.txt') as file:
    input = file.read()
```

```python
prog = intcode.parse(input)
calls = 0
def get(x,y):
    global calls
    calls += 1
    proc = intcode.run(prog.copy())
    next(proc)
    proc.send(x)
    return proc.send(y)
```

```python
%%time

calls = 0
affect = 0
for y in range(50):
    row = ''
    for x in range(50):
        q = get(x,y)
        if q: affect+=1
        row += '⬛⭕'[q]
    print(row)
print('total calls', calls)
affect
```

```python

def getRect(n):
    global calls
    calls = 0
    n-=1
    left = 0
    for y in count(n):
        for x in count(left):
            if get(x,y)==0: continue
            left = x
            if get(x+n,y-n)==0:
                break
            # abusing the shape of the beam.
            print('total calls', calls)
            return x,y-n
```

```python
%%time

getRect(100)
```
